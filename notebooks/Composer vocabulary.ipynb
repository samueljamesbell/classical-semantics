```python
import csv

composers_path = '../data/composers.csv'
composer_names = []
with open(composers_path, 'r') as f:
    rows = csv.reader(f)
    next(rows)  # Skip headers
    for row in rows:
        composer_names.append(row[0])
        
print(composer_names)
        
        
```

```python
name_words = {word for name in composer_names for word in name.split(' ')}
print(name_words)
```

```python
# Write our dataset to CSV
fieldnames = ['name_word']
output_file = '../data/composer_name_words.csv'

with open(output_file, 'w') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()
    
    for name_word in name_words:
        writer.writerow({'name_word': name_word})
        
print('Done!')
```
